```python
import joblib
import pandas as pd

# Load pre-trained model
# (Replace 'diabetes_model.pkl' with your trained model file)
model = joblib.load("diabetes_model.pkl")

print("=" * 50)
print("Disease Prediction System")
print("=" * 50)

# User Input
pregnancies = int(input("Pregnancies: "))
glucose = float(input("Glucose Level: "))
blood_pressure = float(input("Blood Pressure: "))
skin_thickness = float(input("Skin Thickness: "))
insulin = float(input("Insulin Level: "))
bmi = float(input("BMI: "))
diabetes_pedigree = float(input("Diabetes Pedigree Function: "))
age = int(input("Age: "))

# Create DataFrame
patient = pd.DataFrame({
    "Pregnancies": [pregnancies],
    "Glucose": [glucose],
    "BloodPressure": [blood_pressure],
    "SkinThickness": [skin_thickness],
    "Insulin": [insulin],
    "BMI": [bmi],
    "DiabetesPedigreeFunction": [diabetes_pedigree],
    "Age": [age]
})

# Prediction
prediction = model.predict(patient)

# Probability (if supported)
if hasattr(model, "predict_proba"):
    probability = model.predict_proba(patient)[0][1]
else:
    probability = None

# Output
print("\n" + "=" * 50)

if prediction[0] == 1:
    print("Prediction: High Possibility of Diabetes")
else:
    print("Prediction: Low Possibility of Diabetes")

if probability is not None:
    print(f"Confidence: {probability * 100:.2f}%")

print("=" * 50)
```
