```python
# ============================================
# Handwritten Character Recognition using OCR
# ============================================

!pip install easyocr
import easyocr
import cv2
import matplotlib.pyplot as plt
import os

print("=" * 50)
print("Handwritten Character Recognition System")
print("=" * 50)

# Load OCR model (downloads automatically on first run)
print("\nLoading OCR Model...")
reader = easyocr.Reader(['en'])

# Get image path from user
image_path = input("\nEnter the path of the handwritten image: ")

# Check if image exists
if not os.path.exists(image_path):
    print("\nError: Image file not found!")
    exit()

# Read image
image = cv2.imread(image_path)

if image is None:
    print("\nError: Unable to open image.")
    exit()

# Convert image for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display image
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.title("Input Handwritten Image")
plt.axis("off")
plt.show()

# OCR Prediction
results = reader.readtext(image_path)

# Print detected text
print("\nRecognized Text")
print("-" * 40)

if len(results) == 0:
    print("No handwritten text detected.")
else:
    final_text = ""

    for detection in results:
        text = detection[1]
        confidence = detection[2]

        print(f"Text       : {text}")
        print(f"Confidence : {confidence:.2f}")
        print()

        final_text += text + " "

    print("=" * 50)
    print("Final Prediction:")
    print(final_text.strip())
    print("=" * 50)
```
